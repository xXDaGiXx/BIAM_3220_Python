### In the next code cell, create a function called, `dfinfo()` that will show the shape, column names, column datatypes, and first 5 rows of the DataFrame when you call it and pass a DataFrame to it
>- This is something that could save you time when working at a job with a lot of datasets
#### Function Specifications
1. Name the function dfinfo()
2. The function should accept only one parameter, a dataframe object
3. Include a line space in between the shape, columns, datatypes, and header
4. The function should print out the shape, columns and datatypes as well as return the first five records. Include printed text to show what each piece of information is showing

Below is the general layout of what your function should show when called:

    Number of rows and columns: <print (rows,columns)>

    The columns are: <(print the columns)>

    Datatypes are: <(print the datatypes)>

    <After the print statements, return the first 5 rows after the above print statements> 

```python
def dfinfo(d):
    print(f'Number of rows and columns: {d.shape}\n')
    print(f'The columns are: {tuple(d.columns)}\n')
    print(f'Datatypes are: {d.dtypes}\n')
    display(d.head())
```

### Refer to randomly generated question on Canvas (copy and paste it in this markdown cell)
#### For jobs that had total employment between 500,000 and 1,000,000 in 2011, which job category had a percentage growth in employment from 2010 to 2011 equal to -1.16%? Type (or copy and paste) the Occupation Code exactly as it shows in the spreadsheet. Your answer must have the form ##-####.

```python
totEmp10 = bls4yr.loc[bls4yr['year']==2010].groupby('occ_code').tot_emp.sum()

totEmp10
```

```python
totEmp11 = bls4yr.loc[bls4yr['year']==2011].groupby('occ_code').tot_emp.sum()
totEmp11
```

```python
totEmp = round(((totEmp11 - totEmp10)/totEmp10)*100,2)
totEmp.to_frame()
```

```python
totEmp[totEmp == -1.16]
```
