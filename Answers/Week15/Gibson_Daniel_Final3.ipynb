# Question 7
## Show your work to answer question 7 from Canvas here
#### What was the Yearly Rate of Return for Royal Caribbean Cruises (RCL) in 2018? Round your answer to two decimal places
>- Note: Not all days are trading days
>- Yearly Rate of Return = ((End of Year Value - Start of Year Value)/Start of Year Value) *100

```python
rcl['Year'] = rcl.index.year

rcl.tail()
```

```python
rcl1 = rcl.groupby(['Year'], as_index=False).agg('max').round(2)
```

```python
rcl1['annRoR'] = round(((rcl1['Close']-rcl1['Close'].shift(1))/(rcl1['Close'].shift(1)))*100,2)
rcl1.tail()
```
