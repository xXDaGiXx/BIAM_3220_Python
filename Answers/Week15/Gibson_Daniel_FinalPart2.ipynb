#### How did the mean annual wage for Building and grounds cleaning and maintenance occupations rank out of all 22 major occupations in 2009? Rank from highest to lowest where rank = 1 means the occupations with the highest mean annual wage.

#### For jobs that had total employment between 500,000 and 1,000,000 in 2011, which job category had a percentage growth in employment from 2010 to 2011 equal to 0.38%? Type (or copy and paste) the Occupation Code exactly as it shows in the spreadsheet. Your answer must have the form ##-####.

```python
totEmp10 = bls4yr.loc[bls4yr['year']==2010].groupby('occ_code').tot_emp.sum()
totEmp10
```

```python
totEmp11 = bls4yr.loc[bls4yr['year']==2011].groupby('occ_code').tot_emp.sum()
totEmp11
```

```python
totEmp1 = totEmp10+totEmp11
totEmp1
```

```python
totEmp1['45-0000']
```

```python
totEmp1['49-3023']
```

```python
totEmp10 = bls4yr.loc[bls4yr['year']==2010].groupby('occ_code').tot_emp.sum()
totEmp11 = bls4yr.loc[bls4yr['year']==2011].groupby('occ_code').tot_emp.sum()
totEmp = round(((totEmp11 - totEmp10)/totEmp10)*100,2)
totEmp.to_frame()
```
