## Q17 (6 points):
### Task: Create a new dataframe named, `custRev`, based on data in the `ordL` dataframe 
>- `custRev` should only include customer name and the sum of the revenue per customer in descending order
>- Let your index be auto generated and not one of the two fields requested 

```python
custRev = ordL.groupby('customername').revenue.agg(['sum']).sort_values(by = 'sum', ascending = False)
custRev
```

## Q19: (6 Points)
### What is the overall average revenue  from the `custRev` DataFrame?
>- Round to two decimal places
>- Store the result in a variable called `avgRev`
>- Show the output of `avgRev`

```python
# avgRev = round(custRev.sum.mean(),2)
avgRev = round(custRev['sum'].sum()/len(custRev.index),2)
avgRev
```

## Q20 (6 points)
### Feature Engineering Task
#### Create a new column, `avgdiff`, to show the difference in customer revenue and average customer revenue (`avgRev`)
>- Show the first five records of `custRev` after you complete the task

```python
custRev['avgDiff'] = custRev['sum'] - avgRev
custRev.head()
```
