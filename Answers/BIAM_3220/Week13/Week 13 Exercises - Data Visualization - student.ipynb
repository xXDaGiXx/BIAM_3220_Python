### Creating a Line Plot
>- Here, we will make up some data just to get familiar with syntax. Later you will apply these methods to other datasets (BLS, SaleCo, etc)

```python
 # create some variables to plot
x, y = [1600,1700,1800,1900,2000],[0.2,0.5,1.1,2.2,7.7]
```

#### Now, plot the data using a line plot with `plt.plot()`

```python
plt.plot(x,y)
```

#### Now, let's add some features
>- Graph Title with `plt.title()`
>- Axis lables with `plt.xlabel` and `plt.ylabel`
>- Adjust figure size with `plt.figure(figsize=(width, height))`

```python
# Adjust the figure size
plt.figure(figsize=(15,8))

# Add a title to the chart
plt.title('world population over time'.title())

# Add labels to are x and y axis
plt.xlabel('Year')
plt.ylabel('Population (billions)')

plt.plot(x,y)

plt.show()
```

### Plotting multiple series on the same chart

>- Within each plot() call we will pass different settings for the lines and markers
>>- For example `plt.plot(x1, y1, "r-x", label="Actual")`
>>>- This plots x1 and y1 with a red (r) sold line (-) with "x" data markers

```python
# Actual data
x1, y1 = [1600,1700,1800,1900,2000],[0.2,0.5,1.1,2.2,7.7]

# Second dataset, predicted
x2,y2 = [1600,1700,1800,1900,2000],[1,1,2,3,4]

# set the figure size
plt.figure(figsize=(16,9))

# create a red solid line with x dots for the actual dataset
plt.plot(x1,y1, "r-x",label = 'Actual')

# Create a blue dashed line with circle dots
plt.plot(x2,y2,"b--o",label= 'Predicted')

# customize the chart
plt.title("World Pop Over Time")
plt.xlabel('Year')
plt.ylabel('Population (billions)')

# Show labels for each dataset / series in the best corner
plt.legend()

plt.show()
```

## Creating a Bar Plot

##### First, create some data

```python
num_people = [4,8,3,6,2]
categories = ['Comedy','Action','Thriller','Romance','Horror']
```

##### Next, use `plt.bar()` to create a bar chart

```python
plt.bar(categories,num_people)
plt.show()
```

##### Add data labels with `zip()` and `annotate()`

>- `zip()` pairs items items between iterators (lists, tuples, etc)
>- `zip()` Reference: https://www.w3schools.com/python/ref_func_zip.asp
>- `annotate()` Reference: https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.annotate.html

```python
npLab = zip(num_people,categories)
# print(tuple(npLab))
```

#### Now re-create our plot to show data labels

```python
npList = list(npLab)
# print(tuple(npLab))
n,c = zip(*npList)
print(n)
print(c)

plt.bar(categories,num_people)
plt.annotate(c,xy=(n,0))
plt.plot(n,c)
plt.show()
```

## Creating a Boxplot
>- Boxplots are useful in situations where you want to compare a single metric (e.g., prices, units sold, heights, weights, etc) over time or when comparing categories (regions, country, gender, etc)

```python
males = [72,68,65,77,73,71,69]
females = [60,65,68,61,63,64]
plt.figure(figsize=(10,5))
plt.boxplot([males,females],showmeans=True)
plt.xticks([1,2],['Male',"Female"])
plt.title('Height by Gender', fontsize=24)
plt.ylabel('Height (inches)', fontsize = 14)
plt.xlabel('Gender',fontsize=14)
plt.show()
```

## Create a Scatter Plot

##### First, we will create some random data for height and weight

```python
height = [507,508,509,510,511,512,]
weight = [154,125,188,215,97,105]
```

##### Now plot using `plt.scatter()`

```python
plt.figure(figsize=(15,8))
plt.scatter(height,weight)
plt.title('Height vs Weight',fontsize=20)
plt.xlabel('Height (Feet and Inches)',fontsize=12)
plt.ylabel('Weight (lbs)',fontsize=12)
plt.show()
```
